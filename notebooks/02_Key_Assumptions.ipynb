# Theoretical Foundations: Metric Neighborhoods & Set Covers

## Assumption 1
*   **Assumption:** The underlying geometric relationship distinguishing class margins exhibits variations that can be accurately captured within localized metric spaces (such as $k$-neighbourhoods).
*   **Why the method needs it:** The algorithm divides the unified global prediction problem into multiple independent local SVM boundaries. If the classes are heavily intertwined uniformly and local features cannot dictate valid boundaries, then training locally introduces tremendous variance without meaningful separation, collapsing the method's reliability.
*   **Violation scenario:** In ultra-sparse high-dimensional datasets (e.g., TF-IDF embedded natural language text), the assumption of meaningful local geometric neighbourhoods breaks down entirely due to the phenomenon where pairwise points tend to become effectively equidistant, eradicating valid spatial decision gradients.
*   **Paper reference:** Section 1, Introduction (discussion on "intrinsic dimensionality which is low with respect to the training set size" vs "high-dimensionality problems") and Section 4.5.1 ("Curse of Dimensionality").

## Assumption 2
*   **Assumption:** A cohesive $k'$-neighbourhood covering set that represents the whole geometric scope of the dataset can be successfully generated by bounding local overlapping hyperspheres around a limited group of centre nodes $C$.
*   **Why the method needs it:** Step 2 explicitly counts on this bounding condition to select a significantly reduced set of centres $C \subset X$ for which local SVMs will be computed. If a cohesive cover set is geometrically impossible using $k'$, the algorithm will be forced to allocate a centre to nearly every point individually.
*   **Violation scenario:** If the training data distribution severely resembles noise and consists of countless isolated anomalies or minute disjoint clusters completely separated by vast empty metric distances, minimizing $|C|$ will fail aggressively since nearly every anomaly would require its own $k'$-neighbourhood. e.g. a dataset of 1000 points each differing by exactly one feature, like one-hot encoded categories.
*   **Paper reference:** Section 4.2.1 (Selecting the Centres of the Local Models: Definition 1, Definition 2, and the formulation utilizing Set Cover problem principles).

## Assumption 3
*   **Assumption:** The discrete proximity between a testing query and its absolute nearest neighbour in the training dataset serves as an accurate proxy for dictating the optimal maximum-margin hyperplane model without recalculating it dynamically.
*   **Why the method needs it:** Instead of performing a lazy "query-centric" approach like traditional k-Nearest Neighbours or Local SVM, FaLK-SVM arbitrarily adopts the precalculated neighbourhood hyperplane originating directly from the query's nearest Euclidean training point.
*   **Violation scenario:** In regions carrying dense overlapping label distributions (a fuzzy overlapping decision frontier with high local noise), the absolute nearest neighbour measured purely by distance metric might belong to the opposing class outlier, enforcing a structurally flawed localized subset hyperplane choice to evaluate the test distribution.
*   **Paper reference:** Section 4.4 (Generalisation Bounds), wherein the risk penalty term expected value $E[\lambda]$ mathematically addresses the proxy loss deviation derived by adopting the nearest neighbour $x_{rq}(1)$ 's pre-computed parameters over $x'$ intrinsically.
