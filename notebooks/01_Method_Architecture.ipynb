## Step-by-Step Method Description

*   **Step 1: Pre-computing Local Models during Training Phase**
    *   **Description**: Instead of delaying the neighbourhood retrieval and model training until a testing point is available (lazy learning), the method shifts to eager learning. It entails estimating a local Support Vector Machine (SVM) decision function for the $k$-neighbourhood of training points beforehand.
    *   **Reference**: Section 4.1 (Equation 5)
    *   **Purpose**: This transitions the local computing paradigm to an eager learning setting, drastically improving the computational complexity and efficiency of the testing phase.
*   **Step 2: Selecting the Centres of the Local Models**
    *   **Description**: To avoid training an independent local SVM model for every single point in the training set $X$, a subset $C \subseteq X$ is smartly selected. A greedy Set Cover approximation strategy is employed so that every training sample is covered by at least one $k'$-neighbourhood of a centre in $C$.
    *   **Reference**: Section 4.2.1 (Definition 1, Equation 9)
    *   **Purpose**: Reduces the number of requisite local SVMs from $N$ to $|C|$, mitigating the dramatic training overhead while maintaining the accuracy characteristics of local learning spaces.
*   **Step 3: Assigning Local Models for Testing Points**
    *   **Description**: Each training point is uniquely assigned to one of the trained local models centered at a point from the set $C$, chosen based on the minimum neighbourhood rank constraint.
    *   **Reference**: Section 4.2.2 (Equation 11 - `cnt` function)
    *   **Purpose**: Ensures rigorous consistency by linking any future testing queries to the exact local SVM model pre-computed on the corresponding spatial domain.
*   **Step 4: Label Prediction**
    *   **Description**: For a given query geometry, the FastLSVM approach simply retrieves the query's nearest Euclidean neighbor in the training set and leverages its pre-assigned local model to output a margin-based prediction.
    *   **Reference**: Section 4.2.2 (Equation 12 - `FaLK-SVM` prediction strategy)
    *   **Purpose**: Accelerates classification down to logarithmic time complexity $\mathcal{O}(\max(\log N, k))$ using cover trees, providing highly efficient prediction latency suitable for classifying vast streams of unseen instances.
