# Task 1.3 What the Paper Claims to Improve

*   **Main Baseline Method**: The main baselines that the paper compares against are **Local SVM (specifically the kNNSVM formulation)** and **global Support Vector Machine (LibSVM)**.
*   **Limitation Identified**: While global SVM architectures struggle computationally with massive scale problems (failing to execute beyond reasonable timeframes or globally under-fitting via approximations), the primary direct baseline Local SVM (kNNSVM) suffers from dire latency computational costs exclusively during the testing phase. Because it relies heavily on lazy learning, it mandates recursively hunting for the top $k$-neighbours and solving a fresh exact SVM optimization exclusively anew at the very moment each unseen test instance presents.
*   **How Proposed Method Overcomes This**: The authors mitigate this foundational limitation by transitioning from lazy memory learning to an eager execution classifier. FaLK-SVM resolves spatial subsets via redundant $k'$-neighbourhood Set Cover logic to preemptively resolve generalized local SVMs exclusively at subset centres. Testing latency is thereby vastly reduced to mere logarithmic queries retrieving proximity structures internally indexed by cover trees.
*   **Failure Condition Scenario**: Based on principled reasoned inference regarding mechanics, a stark scenario under which FaLK-SVM may categorically fail to outperform the global SVM baseline includes modeling fundamentally globally-linear patterns residing in artificially vast high-dimensional arrays devoid of distinct cluster attributes. Here, navigating multidimensional cover tree hierarchies to query uncalibrated neighbourhoods injects significant topological delays into the system, drastically diminishing throughput vs standard linear conjugate gradients, whilst failing utterly to capture higher testing precision since exact continuous global hyperplanes would otherwise natively perfectly sever the entire sample space instantaneously.
